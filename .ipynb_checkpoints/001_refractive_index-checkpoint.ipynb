```python
x = np.linspace(0.4,0.7,100)
```

```python
ne = 1.6933 + 0.0078*x**(-2) + 0.0028*x**(-4)
no = 1.4990 + 0.0072*x**(-2) + 0.0003*x**(-4)
```

```python
def n_e(lam):
    return 1.6933 + 0.0078*lam**(-2) + 0.0028*lam**(-4)

def n_o(lam):
    return 1.4990 + 0.0072*lam**(-2) + 0.0003*lam**(-4)
```

```python
ne = n_e(x)
no = n_o(x)
```

```python
plt.plot(x,ne,c='blue',label=r'$n_e$')
plt.plot(x,no,c='red',label=r'$n_o$')
plt.axvline(0.55,c='k',ls='--')
plt.grid()
plt.xlabel(r'$\lambda \, [\mu m]$',fontsize=14)
plt.ylabel(r'$n$',fontsize=14)
plt.legend(fontsize='xx-large')
plt.tick_params(axis='both',labelsize=14)
plt.title('E7',fontsize=18)
plt.show()
```

```python
microns = 0.55
print("Extraordinary refractive index at lambda = %.2f: \t ne=%.3f"%(microns,n_e(microns)))
print("Ordinary refractive index at lambda = %.2f: \t\t no=%.3f"%(microns,n_o(microns)))
```

```python
def nx(x,y,z):
    return np.zeros_like(x)
def ny(x,y,z):
    return np.zeros_like(x)
def nz(x,y,z):
    return np.ones_like(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
z = np.linspace(-5, 5, 50)
X, Y, Z = np.meshgrid(x, y, z)

# Create scalar field (example: 3D Gaussian)
field = np.exp(-(X**2 + Y**2 + Z**2)/10)

# Slice visualization
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot isosurfaces (using contourf on slices)
slice_val = 0
c = ax.contourf(X[:, :, 25], Y[:, :, 25], field[:, :, 25], zdir='z', offset=0, cmap='viridis')
plt.colorbar(c)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Slice of 3D Scalar Field')
plt.show()
```

```python
for i in range(40):
    for j in range(40):
        plt.plot(data['u'][i][j],label='u')
        plt.plot(data['v'][i][j],label='v')
        plt.plot(data['w'][i][j],label='w')
plt.ylabel('director field')
plt.xlabel('z axis')
#plt.legend()
plt.show()
```

```python
# the first two indexes are for x and y
# I chose (0,0) but all the other pixels in the xy plane have the same value 
# you can try by changing  (0,0) into any combination of numbers (0-39,0-39)
# the trend on the z axis (that is what we are looking in this plot is the same for every pixel)
# the idea is that we are chaning the polarization in the (xy plane) as a function of the z axis

# the loop is on the 10 element in the z axis
for k in range(10):
        plt.plot(k,data['u'][0][0][k],"ro",label='u')
        plt.plot(k,data['v'][0][0][k],"go",label='v')
        plt.plot(k,data['w'][0][0][k],"bo",label='w')
plt.ylabel('director field')
plt.xlabel('z axis')
#plt.legend()
plt.show()
```
